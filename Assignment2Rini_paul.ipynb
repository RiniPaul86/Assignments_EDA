#IPL 2008 to 2022 (Match Details)

######IPL is among the most famous cricket league with players coming from worldwide. What makes this series more competitive every year is the drama with every ball and hence this dataset is a complete dataset showing match information and also data for each delivery.

The file IPL_Matches_2008_2021.csv contains all details of the match: Venue of the match, Toss Decision, Match Winer, Man of the Match, Squads, etc.


```python
#load the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data1 = pd.read_csv("/content/IPL_Matches_2008_2022.csv")
data1.head()
```

```python
#dropping the columns
data1 = data1.drop(['ID','Date','MatchNumber','SuperOver','method','Team1Players','Team2Players'],axis = 1)
```

####Mean Imputation

```python
data1.isnull().sum()
```

```python
med1 = data1['Margin'].median()
print(med1)
```

```python
data1['Margin'] = data1['Margin'].fillna(med1)
```

```python
data1.isnull().sum()
```

```python
data1.describe()
```

```python
#Box Plot
data1 = data1.dropna()
```

```python
plt.boxplot(data1.Margin)
```

```python
#Filtering
data1_sub1 = data1[data1['Margin']>100]
data1_sub1.size
```

```python
data1_sub2 = data1[data1['Margin']<30]
data1_sub2.size
```

```python
data1_sub3 = data1[data1['Margin']>140]
data1_sub3.size
```

```python
#slicing
data1[['TossWinner','TossDecision']].head(10)
```
