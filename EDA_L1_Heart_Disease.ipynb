```python
# Creating a new DataFrame with changed numerical values to categorical values for various analysis purposes.
df_new = df.copy()
```

```python
# for better insights changed numerical values to categorical values for certain columns.
df_new['sex '] = df_new['sex '].map({1: 'male', 0: 'female'})
df_new['heart disease'] = df_new['heart disease'].map({1: 'Absent', 2: 'Present'})
df_new['chest pain type'] = df_new['chest pain type'].map({1: 'typical angina', 2: 'atypical angina', 3: 'non-anginal', 4: 'asymptomatic'})
df_new['thal'] = df_new['thal'].map({3: 'normal', 6: 'fixed defect', 7: 'reversable defect'})
```

# Max and Mini Resting Blood Pressure wrt Heart disease

```python
# Filter dataset for heart disease present and absent cases
heart_disease_present = df_new[df_new['heart disease'] == 'Present']
heart_disease_absent = df_new[df_new['heart disease'] == 'Absent']
```

```python
# Find the highest and lowest resting blood pressure values among cases of heart disease present
max_bp_present = heart_disease_present['resting blood pressure'].max()
min_bp_present = heart_disease_present['resting blood pressure'].min()

# Find the highest and lowest resting blood pressure values among cases of heart disease absent
max_bp_absent = heart_disease_absent['resting blood pressure'].max()
min_bp_absent = heart_disease_absent['resting blood pressure'].min()

# Print the results
print("Highest Resting Blood Pressure among Heart Disease Present:", max_bp_present)
print("Lowest Resting Blood Pressure among Heart Disease Present:", min_bp_present)
print("Highest Resting Blood Pressure among Heart Disease Absent:", max_bp_absent)
print("Lowest Resting Blood Pressure among Heart Disease Absent:", min_bp_absent)
```

###The range of resting blood pressure values is wider among individuals with heart disease (100 mmHg to 200 mmHg) compared to those without heart disease (94 mmHg to 180 mmHg).

###The highest recorded resting blood pressure among individuals with heart disease (200 mmHg) indicates the presence of hypertension in some cases, which is a significant risk factor for heart disease.

###Similarly, even among individuals without heart disease, the recorded resting blood pressure values can indicate potential hypertension, which is important for early intervention and prevention of heart disease.

3. ST Segment
