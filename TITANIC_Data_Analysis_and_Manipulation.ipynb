The Titanic dataset is a well-known dataset in the field of data science and machine learning. It contains information about the passengers aboard the RMS Titanic, which sank on its maiden voyage in April 1912 after colliding with an iceberg. :
The RMS Titanic was a British passenger liner that sank on its maiden voyage from Southampton to New York City. It struck an iceberg on the night of April 14, 1912, resulting in the loss of over 1,500 lives out of the approximately 2,224 passengers and crew aboard.


---




---


The Titanic dataset provides information about each passenger, including:

PassengerId: Unique identifier for each passenger.

Survived: Indicates whether the passenger survived (0 = No, 1 = Yes).

Pclass: Ticket class (1st = Upper, 2nd = Middle, 3rd = Lower).

Name: Passenger's name.

Sex: Passenger's gender.

Age: Passenger's age.

SibSp: Number of siblings/spouses aboard.

Parch: Number of parents/children aboard.

Ticket: Ticket number.

Fare: Passenger fare.

Cabin: Cabin number.

Embarked: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton).


---



---

The dataset is often used for predictive modeling to answer questions like "What factors made people more likely to survive?" This involves building machine learning models to predict whether a passenger survived based on the other variables in the dataset.

#1. Implementing Basic Array Operations:

a. Perform addition operation on the 'Fare' column of the Titanic dataset. What is the total fare paid by all passengers?

b. Subtract the 'Age' column from the 'Fare' column. What insights can you derive from this operation?

c. Multiply the 'SibSp' column (number of siblings/spouses aboard) with the 'Parch' column (number of parents/children aboard). What does the resulting column represent?

d. Divide the 'Fare' column by the 'Pclass' column (passenger class). Interpret the meaning of the values obtained.

```python
# Perform addition on the 'Fare' column
total_fare = df['Fare'].sum()
```

```python
print("Total Fare Paid by all Passengers: $", total_fare)
```

```python
# Subtract 'Age' from 'Fare'
df['Fare_Minus_Age'] = df['Fare'] - df['Age']
```

```python
# Display the first few rows to see the result
print(df[['Fare', 'Age', 'Fare_Minus_Age']].head())
```

Analyzing this new column can provide insights into potential data issues, unusual patterns, or outliers in the dataset. It's important to further investigate any unexpected results to ensure data integrity and accuracy.

```python
# Multiply 'SibSp' with 'Parch'
df['SibSp_Parch_Product'] = df['SibSp'] * df['Parch']
```

```python
# Display the first few rows to see the result
print(df[['SibSp', 'Parch', 'SibSp_Parch_Product']].head())
```

The resulting 'SibSp_Parch_Product' column represents the number of family members a passenger has aboard the Titanic.

Larger values in 'SibSp_Parch_Product' indicate passengers who are traveling with more family members, while smaller values indicate passengers who are traveling alone or with fewer family members.

```python
# Divide 'Fare' by 'Pclass'
df['Fare_Per_Pclass'] = df['Fare'] / df['Pclass']
```

```python
# Display the first few rows to see the result
print(df[['Fare', 'Pclass', 'Fare_Per_Pclass']].head())
```

Higher values in 'Fare_Per_Pclass' indicate passengers who paid more for their class.

Lower values in 'Fare_Per_Pclass' indicate passengers who paid less for their class.



#2. Creating and Manipulating Multidimensional Arrays:

a. Create a 2D array using the 'Age' and 'Fare' columns of the Titanic dataset. What are thedimensions of this array?

b. Perform element-wise multiplication on this array with a scalar value of 10. What effect does this operation have on the array?

c. Transpose the array created in part (a). What are the dimensions of the transposed array?

```python
# Select 'Age' and 'Fare' columns
selected_data = df[['Age', 'Fare']]

# Convert selected data to a 2D array
data_array = selected_data.to_numpy()

data_array
```

```python
# Get the dimensions of the array
array_shape = data_array.shape

print("2D Array Dimensions (Rows, Columns):", array_shape)
```

```python
# Perform element-wise multiplication with scalar (10)
multiplied_array = data_array * 10
```

```python
# Print the original array and the multiplied array for comparison
print("Original 2D Array:")
print(data_array[:5])  # Print first 5 rows for illustration
print("\nArray after Element-wise Multiplication with Scalar (10):")
print(multiplied_array[:5])  # Print first 5 rows for illustration
```

The effect of this operation is that all values in the array are now 10 times larger than before.

```python
# Transpose the multiplied array
transposed_array = multiplied_array.T

# Get the dimensions of the transposed array
transposed_shape = transposed_array.shape

print("\nDimensions of Transposed Array (Columns, Rows):", transposed_shape)
```

#6. Data Manipulation Tasks:
a. Filter the Titanic dataset to include only passengers with ages between 20 and 30 years.

b. Sort the dataset based on the 'Fare' column in descending order.

c. Group the dataset by 'Pclass' (passenger class) and calculate the mean fare for each class.

```python
filtered_data = df[(df['Age'] >= 20) & (df['Age'] <= 30)]

# Display the filtered dataset
print("Passengers with ages between 20 and 30:")
filtered_data.head()
```

```python
# Sort the dataset based on 'Fare' column in descending order
sorted_df = df.sort_values(by='Fare', ascending=False)

# Display the sorted dataset
print("Dataset Sorted by 'Fare' in Descending Order:")
print(sorted_df[['Name', 'Fare']].head(10))  # Displaying only 'Name' and 'Fare' for the top 10
```

```python
# Group the dataset by 'Pclass' and calculate mean fare
mean_fare_by_class = df.groupby('Pclass')['Fare'].mean().reset_index()

# Display the mean fare for each class

print("Mean Fare for Each Passenger Class:")
print(mean_fare_by_class)
```
